```
import numpy as np
import matplotlib.pyplot as plt
```

```
second_octave = np.array([65.41, 73.42, 82.41, 87.31, 98.00, 110.00, 123.47])
third_octave = 2*second_octave
fourth_octave = 2*third_octave
fifth_octave = 2*fourth_octave

full_scale = np.concatenate((second_octave,third_octave,fourth_octave,fifth_octave))

print(full_scale)
```

```
sampling_rate = 10000.
fft_size = 2048

frequency_bins = np.array([x*sampling_rate/fft_size for x in range(fft_size // 2)])

print(frequency_bins)
print(len(frequency_bins))
```

```
lut = [0]*(fft_size // 2)

for index, val in enumerate(frequency_bins):
  abs_array = abs(full_scale - val)
  closest_element_index = abs_array.argmin()
  if (full_scale[-1] - val) < 0:
    lut[index] = full_scale[-1]
  else:
    lut[index] = full_scale[closest_element_index]

print(lut)
print(len(lut))
print(lut[150])
```

```
lut2 = [0]*(fft_size // 2)

for index, val in enumerate(lut):
  ratio = 440. / val
  fcw = 224 / ratio
  lut2[index] = int(fcw)

print(lut2)
print(lut2[150])
```

```
for val in lut2:
  print(hex(val)[2:])
```
